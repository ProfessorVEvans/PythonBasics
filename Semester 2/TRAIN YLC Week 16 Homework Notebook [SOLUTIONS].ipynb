### **Part #1.1: Simple Neural Networks**
---

####**Problem 1.1.1**

Initialize a neural network using `Sequential()` in keras. Add the following layers to the network.

* An input layer with an input size of 10.
* Two hidden layers with the following specifications:
 * The first hidden layer should have 12 neurons.
 * The second hidden layer should have 8 neurons.
* Lastly, add an output layer that has 2 output nodes.

This network can be used for multi-class classification tasks or other suitable applications.

#####**Solution**

```python
# Build the model
model = Sequential()

# Input Layer
model.add(Input(10))

# Hidden Layers
model.add(Dense(12))
model.add(Dense(8))

# Output layer
model.add(Dense(2))
```

#### **Problem #1.1.2**

Confirm we've made our model correctly using the `visualizer` function.



#####**Solution**

```python
# Visualizing the network
visualizer(model, file_name="model",file_format='png')
Image("model.png")
```

####**Problem 1.1.3**

Initialize a neural network using `Sequential()` in keras. Add the following layers to the network.

* An input layer with an input size of 6.
* Two hidden layers with the following specifications:
 * The first hidden layer should have 15 neurons.
 * The second hidden layer should have 10 neurons.
* Lastly, add an output layer that has 3 output nodes.

This type of network might be used for tasks like multi-class classification for three classes.

Finally, visualize the model using the `visualizer` function.

#####**Solution**

```python
# Build the model
model = Sequential()

# Input Layer
model.add(Input(6))

# Hidden Layers
model.add(Dense(15))
model.add(Dense(10))

# Output layer
model.add(Dense(3))

# Visualizing the network
visualizer(model, file_name="model",file_format='png')
Image("model.png")
```

### **Part #1.2: Adding Activation Functions**
---

####**Problem 1.2.1**

Create and visualize a neural network model with the following specifications using keras:

* An input layer with 3 neurons.
* A single hidden layer with 7 neurons, using the "ReLU" activation function.
* An output layer with 2 neurons, followed by the "ReLU" activation function.

#####**Solution**

```python
# Build the model
model = Sequential()

# Input Layer
model.add(Input(3))

# Hidden Layers
model.add(Dense(7, activation = 'relu'))

# Output layer
model.add(Dense(2, activation = 'relu'))

# Visualizing the network
visualizer(model, file_name="model",file_format='png')
Image("model.png")
```

####**Problem 1.2.2**

Create and visualize a neural network model with the following specifications using keras:

* An input layer with 6 neurons.
* Two hidden layers:
 * The first hidden layer with 8 neurons, using the "Sigmoid" activation function.
 * The second hidden layer with 5 neurons, using the "Tanh" activation function.
* An output layer with 4 neurons, followed by the "ReLU" activation function.

#####**Solution**

```python
# Build the model
model = Sequential()

# Input Layer
model.add(Input(6))

# Hidden Layers
model.add(Dense(8, activation = 'sigmoid'))
model.add(Dense(5, activation = 'tanh'))

# Output layer
model.add(Dense(4, activation = 'relu'))

# Visualizing the network
visualizer(model, file_name="model",file_format='png')
Image("model.png")
```
